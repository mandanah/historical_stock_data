```python
import requests 
import pandas as pd
import numpy as np
import datetime
from bokeh.plotting import figure, output_file, show
from bokeh.resources import CDN
from bokeh.embed import file_html, components
```

```python
r = requests.get('https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=IBM&outputsize=full&apikey=4W5ULX1BFQRP9E52')
```

```python
all_data = r.json()
```

```python
all_data['Time Series (Daily)']
```

```python
df = pd.DataFrame(all_data['Time Series (Daily)']).transpose()
df.index = pd.to_datetime(df.index)
df
```

```python
def data_for_requested_timeperiod(df, start_date, end_date):
    temp_df = df.drop(['1. open', '4. close', '5. volume'], axis=1)
    temp_df = temp_df[(df.index>= start_date) & (df.index<= end_date)]
    return temp_df
```

```python
ans = data_for_requested_timeperiod(df, '2019-08-13','2020-08-16')
ans
```

```python
def plotting(df):
    p = figure(plot_width=400, plot_height=400, x_axis_type="datetime")
    p.vbar(x = df.index, width=1, bottom = ans['3. low'], top=ans['2. high'], color="firebrick")
    script, div = components(p)
    return script, div
```
